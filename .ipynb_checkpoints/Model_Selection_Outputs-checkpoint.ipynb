```python
import itertools
from collections import defaultdict, OrderedDict
import matplotlib.pyplot as plt
# OR_sus=1.541
# OR_sev=2.623
# OR_ben=4.6
# OR_bar=0.762
# OR0=0.35

OR_sus=1.11
OR_sev=1.62
OR_ben=1.72
OR_bar=0.86
OR0=0.35
```

```python
def calculate_probability(x0, x1, x2, x3):
    p = (OR0 * OR_sus**x0 * OR_sev**x1 * OR_ben**x2 * OR_bar**x3) / (1 + (OR0 * OR_sus**x0 * OR_sev**x1 * OR_ben**x2 * OR_bar**x3))
    return p
```

```python
possible_variable_effects = set(itertools.product(["EO", "ER"], repeat=4)) - set([('ER', 'ER', 'ER', 'ER'), ('EO', 'EO', 'EO', 'EO')])
print(possible_variable_effects)
```

```python
dict = {
    0: "susceptibility",
    1: "severity",
    2: "benefits",
    3: "barriers"
}
count = 0
perm_count = 0
mappings_count = defaultdict(int)
for comb in possible_variable_effects:
    num_education = comb.count("EO")
    num_income = comb.count("ER")
    education_mappings = itertools.product(itertools.permutations([0, 1]), repeat=num_education)
    income_mappings = itertools.product(itertools.permutations([0, 1]), repeat=num_income)
    combined_mappings = itertools.product(education_mappings, income_mappings)
    mapping_set = set()
    for mapping in combined_mappings:
        i_edu = 0
        i_inc = 0
        x_low_education_low_income = []
        x_high_education_low_income = []
        x_low_education_high_income = []
        x_high_education_high_income = []
        final_mapping = []
        
        
        for i in range(len(comb)):
            if comb[i] == "EO":
                x = mapping[0][i_edu]
                x_low_education_low_income.append(x[0])
                x_high_education_low_income.append(x[1])
                x_low_education_high_income.append(x[0])
                x_high_education_high_income.append(x[1])
                i_edu += 1
            else:
                x = mapping[1][i_inc]
                x_low_education_low_income.append(x[0])
                x_high_education_low_income.append(x[0])
                x_low_education_high_income.append(x[1])
                x_high_education_high_income.append(x[1])
            final_mapping.append(x)
                
                
        low_education_low_income = calculate_probability(x_low_education_low_income[0], x_low_education_low_income[1], x_low_education_low_income[2], x_low_education_low_income[3])
        high_education_low_income = calculate_probability(x_high_education_low_income[0], x_high_education_low_income[1], x_high_education_low_income[2], x_high_education_low_income[3])
        high_education_high_income = calculate_probability(x_high_education_high_income[0], x_high_education_high_income[1], x_high_education_high_income[2], x_high_education_high_income[3])
        low_education_high_income = calculate_probability(x_low_education_high_income[0], x_low_education_high_income[1], x_low_education_high_income[2], x_low_education_high_income[3])        
        perm_count += 1
        if high_education_low_income >= high_education_high_income >= low_education_low_income >= low_education_high_income:
            final_mapping = tuple(final_mapping)
            if final_mapping not in mapping_set:
                count += 1
                mapping_set.add(final_mapping)
                print(comb, final_mapping)
                print(high_education_low_income, high_education_high_income, low_education_low_income, low_education_high_income)
                for i in range(len(comb)):
                    level = "high" if final_mapping[i][1] == 1 else "low"
                    mapping_str = "High " + comb[i] + ": " + level + " " + dict[i]
                    print(mapping_str)
                    mappings_count[mapping_str] += 1
                print("\n")
print(count)
print(perm_count)
```

```python
mappings_count = OrderedDict(sorted(mappings_count.items(), key=lambda item: item[1]))
plt.xticks(rotation='vertical')
plt.bar(mappings_count.keys(), mappings_count.values())
```

# Case when education and income can both affect

```python
def combine_x(edu, inc):
    if edu == -1 and inc == -1:
        return 0
    else:
        return max(edu, inc)
```

```python
dict = {
    0: "susceptibility",
    1: "severity",
    2: "benefits",
    3: "barriers"
}
count = 0
perm_count = 0

education_mappings = itertools.product([(0, 1), (1, 0), (-1, -1)], repeat=4)
income_mappings = itertools.product([(0, 1), (1, 0), (-1, -1)], repeat=4)
combined_mappings = itertools.product(education_mappings, income_mappings)
mapping_set = set()
mappings_count_edu = defaultdict(int)
mappings_count_inc = defaultdict(int)
for mapping in combined_mappings:
    x_low_education_low_income = []
    x_high_education_low_income = []
    x_low_education_high_income = []
    x_high_education_high_income = []
    
    edu_mapping = mapping[0]
    inc_mapping = mapping[1]

    for i in range(4):
        x_low_edu = edu_mapping[i][0]
        x_high_edu = edu_mapping[i][1]
        x_low_inc = inc_mapping[i][0]
        x_high_inc = inc_mapping[i][1]
        
        x_low_education_low_income.append(combine_x(x_low_edu, x_low_inc))
        x_high_education_low_income.append(combine_x(x_high_edu, x_low_inc))
        x_low_education_high_income.append(combine_x(x_low_edu, x_high_inc))
        x_high_education_high_income.append(combine_x(x_high_edu, x_high_inc))


    low_education_low_income = calculate_probability(x_low_education_low_income[0], x_low_education_low_income[1], x_low_education_low_income[2], x_low_education_low_income[3])
    high_education_low_income = calculate_probability(x_high_education_low_income[0], x_high_education_low_income[1], x_high_education_low_income[2], x_high_education_low_income[3])
    high_education_high_income = calculate_probability(x_high_education_high_income[0], x_high_education_high_income[1], x_high_education_high_income[2], x_high_education_high_income[3])
    low_education_high_income = calculate_probability(x_low_education_high_income[0], x_low_education_high_income[1], x_low_education_high_income[2], x_low_education_high_income[3])        
    perm_count += 1
    if high_education_low_income >= high_education_high_income >= low_education_low_income >= low_education_high_income:
        if mapping not in mapping_set:
            count += 1
            mapping_set.add(mapping)
            print(mapping)
            print(high_education_low_income, high_education_high_income, low_education_low_income, low_education_high_income)
            for i in range(4):
                if mapping[0][i][1] == 1:
                    level_edu = "high"
                elif mapping[0][i][1] == 0:
                    level_edu = "low"
                else:
                    level_edu = "no effect on"
                    
                if mapping[1][i][1] == 1:
                    level_inc = "high"
                elif mapping[1][i][1] == 0:
                    level_inc = "low"
                else:
                    level_inc = "no effect on"
                    
                mapping_str_edu = "High EO: " + level_edu + " " + dict[i]
                mapping_str_inc = "High ER: " + level_inc + " " + dict[i]
                print(mapping_str_edu)
                print(mapping_str_inc)
                mappings_count_edu[mapping_str_edu] += 1
                mappings_count_inc[mapping_str_inc] += 1
                print("\n")
                

print(count)
print(perm_count)
```

```python
mappings_count_edu = OrderedDict(sorted(mappings_count_edu.items(), key=lambda item: item[1]))
plt.xticks(rotation='vertical')
plt.bar(mappings_count_edu.keys(), mappings_count_edu.values())
```

```python
mappings_count_inc = OrderedDict(sorted(mappings_count_inc.items(), key=lambda item: item[1]))
plt.xticks(rotation='vertical')
plt.bar(mappings_count_inc.keys(), mappings_count_inc.values())
```
